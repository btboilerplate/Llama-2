```python
from datasets import load_dataset
ds = load_dataset("timdettmers/openassistant-guanaco")
```

```python
ds
```

```python
def format_llama(example):
    text = example['text']

    if "### Human:" in text and "### Assistant:" in text:
        user_part = text.split("### Human:")[1].split("### Assistant:")[0].strip()
        assistant_part = text.split("### Assistant:")[1].strip()

        formatted = f"<s>[INST] {user_part} [/INST] {assistant_part} </s>"
    else:
        formatted = ""

    return {"text": formatted}
```

```python
formatted_ds = ds["train"].map(format_llama)
```

```python
formatted_ds = formatted_ds.filter(lambda x: len(x["text"]) > 0)
```
